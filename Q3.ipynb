```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
```

```python
# Load Dataset
digits = load_digits()
```

```python
# Dataset Information

print("Dataset Shape:", digits.data.shape)
print("Image Shape:", digits.images.shape)
print("Target Labels:", digits.target_names)
```

```python
# Display the first image
plt.figure(figsize=(4,4))
plt.imshow(digits.images[5], cmap='grey')
plt.title(f"Label: {digits.target[0]}")
plt.axis('off')
plt.show()
```

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
digits =load_digits()

fig, axes = plt.subplots(5, 2, figsize=(10, 5))
for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap='gray')
    ax.set_title(f"Digit: {digits.target[i]}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```
