```python
import matplotlib.pyplot as plt

# Data for line plot
epochs = [1, 2, 3, 4, 5]
accuracy = [60, 68, 75, 82, 90]

# Data for bar chart
hardware = ["CPU", "GPU", "TPU"]
fps = [25, 80, 150]

# Create subplots
plt.figure(figsize=(10, 4))

# Left subplot
plt.subplot(1, 2, 1)
plt.plot(epochs, accuracy, marker='o')
plt.title("Training Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy (%)")

# Right subplot
plt.subplot(1, 2, 2)
plt.bar(hardware, fps)
plt.title("Inference Speed")
plt.xlabel("Hardware")
plt.ylabel("Frames Per Second (FPS)")

plt.tight_layout()
plt.show()
```
